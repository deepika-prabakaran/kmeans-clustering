```python
df.loc[df['employment']==1,'employment']='full time'
df.loc[df['employment']==2,'employment']='part time'
df.loc[df['employment']==3,'employment']='no'
df.loc[df['employment']==4,'employment']='other'

plt.figure(figsize=(14,4))
plt.subplot(1,3,1)
eatout=sns.countplot(df['eating_out'])
eatout.set_xticklabels(["never",'1-2','2-3','3-5','everyday'])

plt.subplot(1,3,2)
eatout1=sns.countplot(df['eating_out'], hue=df['employment'])
eatout1.set_xticklabels(["never",'1-2','2-3','3-5','everyday'])
plt.title('by employment status')

plt.subplot(1,3,3)
eatout2=sns.countplot(df['eating_out'], hue=df['Gender'])
eatout2.set_xticklabels(["never",'1-2','2-3','3-5','everyday'])
plt.title('by gender')

plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
X=df[['cuisine','eating_out']].values
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    #preventing random initialization: 'init=k-means++'
    kmeans= KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,11),wcss,color='blue', linestyle='dashed', marker='o',
          markersize=10)
plt.title('Elbow Method', size=20)
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# optimum number of clusters: 3
kmeans = KMeans(n_clusters = 3, init = "k-means++", random_state = 42)
y_kmeans = kmeans.fit_predict(X)
y_kmeans
```

```python
df['cluster']=y_kmeans
df.head()
```

```python
df0 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
plt.scatter(df0.cuisine,df0.eating_out,color='green')
plt.scatter(df2.cuisine,df2.eating_out,color='red')
plt.scatter(df3.cuisine,df3.eating_out,color='black')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.xlabel('cuisine')
plt.ylabel('eating_out')
plt.legend()
```
